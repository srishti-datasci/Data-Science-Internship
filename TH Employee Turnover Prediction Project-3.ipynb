## Train Test Split

```python
X=data.drop('Attrition',axis=1)
y=data['Attrition']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.20,random_state=1234)
```

```python
print("Shape of X_train is ",X_train.shape)
print("Shape of X_test is ",X_test.shape)
print("Shape of y_train is ",y_train.shape)
print("Shape of y_test is ",X_test.shape)
```

## Decision Tree Classifier

```python
from sklearn.tree import DecisionTreeClassifier
dtree=DecisionTreeClassifier()
dtree.get_params()
```

```python
from sklearn.model_selection import GridSearchCV
param_grid={'criterion':['gini','entropy'],
           'max_depth':[3,4,5,6,7,8],
           'min_samples_leaf': [1,2,3,4],
           'min_samples_split': [2,3,4,5],
           'random_state':[0,42]}
grid_search=GridSearchCV(dtree,param_grid,cv=4,verbose=True)
grid_search
```

```python
gv_search=grid_search.fit(X_train,y_train)
```

```python
gv_search
```

```python
gv_search.best_estimator_
```

```python
gv_search.best_params_
```

```python
gv_search.best_score_
```

```python
dtree=DecisionTreeClassifier(criterion='entropy',max_depth= 3,
             min_samples_leaf=1,min_samples_split= 2,random_state=0)
dtree.fit(X_train,y_train)
```

```python
y_test_predict=dtree.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,f1_score,precision_score,recall_score,roc_curve
dtree_acc_sc=round(accuracy_score(y_test,y_test_predict)*100,2)
dtree_f1_sc=round(f1_score(y_test,y_test_predict)*100,2)
dtree_pre_sc=round(precision_score(y_test,y_test_predict)*100,2)
dtree_recall_sc=round(recall_score(y_test,y_test_predict)*100,2)
print("Accuracy score is",dtree_acc_sc)
print("F1 score score is",dtree_f1_sc)
print("Precision score is",dtree_pre_sc)
print("Recall score is",dtree_recall_sc)
```

```python
from sklearn.metrics import ConfusionMatrixDisplay,confusion_matrix
cmt=confusion_matrix(y_test,y_test_predict)
tn,fp,fn,tp=cmt.ravel()
tn,fp,fn,tp
```

```python
disp=ConfusionMatrixDisplay(confusion_matrix=cmt,display_labels=[False,True])
disp.plot()
plt.grid(False)
```

```python
dtree_predict_proba=dtree.predict_proba(X_test)[:,1]
fpr,tpr,threshold=roc_curve(y_test,dtree_predict_proba)
plt.plot(fpr,tpr)
plt.show()
```
